```python
import FinanceDataReader as fdr
import seaborn as sns
import pandas as pd
import numpy as np


```

```python
#금 선물 가격 들고오기
df = fdr.DataReader('ZG', '2009-01-01') # 금 선물 (ICE)

#코스피 지수 들고오기
kospi = fdr.DataReader('KS11', '2009-01-01')

#S&P500 지표 들고오기
sp500 = fdr.DataReader('US500', '2009-01-01') # NASDAQCOM 나스닥종합지수

#나스닥지표
nd = fdr.DataReader("IXIC", '2009-01-01')

#한국 1년 채권 수익률
KR1YT = fdr.DataReader('KR1YT=RR', '2009-01-01')
#한국 10년 채권 수익률
KR10YT = fdr.DataReader('KR10YT=RR', '2009-01-01')

#미국 1년 만기 국채 수익률
US1YT = fdr.DataReader('US1YT=X', '2009-01-01')

#미국 10년 만기 국채 수익률
US10YT = fdr.DataReader('US10YT=X', '2009-01-01')
#원달러 환율
KRWUSD = fdr.DataReader('USD/KRW', '2009')
# matplotlib and seaborn are used for plotting graphs
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
#금리, 금값, 미국채권, 
```

```python
kospi = kospi[["Close"]]
kospi
```

```python
kospi = kospi[['Close']].rename(columns={"Close":"kospi"})
sp500 = sp500[['Close']].rename(columns={"Close":"sp500"})
nd = nd[['Close']].rename(columns={"Close":"NASDAQ"})
KR1YT = KR1YT[['Close']].rename(columns={"Close":"KR1YT"})
KR10YT = KR10YT[['Close']].rename(columns={"Close":"KR10YT"})
US1YT = US1YT[['Close']].rename(columns={"Close":"US1YT"})
US10YT = US10YT[['Close']].rename(columns={"Close":"US10YT"})
KRWUSD = KRWUSD[['Close']].rename(columns={"Close":"KRWUSD"})
gold = df[['Close']].rename(columns={"Close":"gold"})
```

```python

kospi_5days_mean = kospi[['kospi']].rolling(5).mean().rename(columns={"kospi":"kospi_line5"})
sp500_5days_mean = sp500[['sp500']].rolling(5).mean().rename(columns={"sp500":"sp500_line5"})
nd_5days_mean = nd[['NASDAQ']].rolling(5).mean().rename(columns={"NASDAQ":"NASDAQ_line5"})
KR1YT_5days_mean = KR1YT[['KR1YT']].rolling(5).mean().rename(columns={"KR1YT":"KR1YT_line5"})
KR10YT_5days_mean = KR10YT[['KR10YT']].rolling(5).mean().rename(columns={"KR10YT":"KR10YT_line5"})
US1YT_5days_mean = US1YT[['US1YT']].rolling(5).mean().rename(columns={"US1YT":"US1YT_line5"})
KRWUSD_5days_mean = KRWUSD[['KRWUSD']].rolling(5).mean().rename(columns={"KRWUSD":"KRWUSD_line5"})
gold_5days_mean = gold[['gold']].rolling(5).mean().rename(columns={"gold":"gold_line5"})
```

```python
total = pd.DataFrame()
```

```python
total = pd.concat([total,kospi,sp500,nd,KR1YT,KR10YT,US1YT,US10YT,KRWUSD,gold,kospi_5days_mean,sp500_5days_mean,nd_5days_mean,KR1YT_5days_mean,KR10YT_5days_mean,US1YT_5days_mean,KRWUSD_5days_mean,gold_5days_mean],axis=1)
#total.columns=['kospi','sp500','NASDAQ','KR1YT','KR10YT','US1YT','US10YT','KRWUSD','gold']
total= total[1:]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(features)
scaled_features = scaler.transform(features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

train_x, test_x, train_y, test_y = train_test_split(features, target, test_size=0.19, random_state=0)
```

ExtratreesRegression을 활용하여 만들어 보고싶은 마음이 생김

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
ET_model = ExtraTreesRegressor(random_state = 0).fit(train_x,train_y)
```

```python
ET_predicted = ET_model.predict(test_x)
ET_RMSE = np.sqrt(mean_squared_error(test_y, ET_predicted))
ET_MSE = mean_squared_error(test_y, ET_predicted)
```

```python
print({ET_RMSE},{ET_MSE}) 
```

```python
ET_stock_value = ET_model.predict(features.tail(1))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(random_state = 0).fit(train_x,train_y)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state=0).fit(train_x,train_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predicted = rf_model.predict(test_x)
RMSE = np.sqrt(mean_squared_error(test_y, predicted))
MSE = mean_squared_error(test_y, predicted)
```

```python
print({RMSE},{MSE}) 
```

```python
RF_stock_value = rf_model.predict(features.tail(1))
print(f'오늘의 코스피 예측 지수: {RF_stock_value}')
```

```python
import lightgbm as lgb
train_ds = lgb.Dataset(train_x, label = train_y)
test_ds = lgb.Dataset(test_x, label= test_y)
params = {'learning_rate':0.01,
          'num_iterations': 2500,
          'max_depth':20,
          'boosting' : 'gbdt',
          'objective' : 'regression',
          'metric': 'mse',
          'is_training_metric':True,
          'num_leaves':200,
          'feature_fraction':0.9,
          'bagging_fraction':0.7,
          'bagging_freq':5,
          'seed':2020
          }
lgbm_model = lgb.train(params, train_ds, 1000, test_ds, verbose_eval=100, early_stopping_rounds= 100)
lg_pred = lgbm_model.predict(test_x)

LG_RMSE = np.sqrt(mean_squared_error(test_y, lg_pred))
LG_MSE = mean_squared_error(test_y, lg_pred)
print({LG_RMSE},{LG_MSE}) 
```

```python
LG_stock_value = lgbm_model.predict(features.tail(1))
print(f'오늘의 코스피 예측 지수: {LG_stock_value}')
```
